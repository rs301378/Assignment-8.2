We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset
given below. <br>
Task 1:<br></b>
1.fitting it to the periodic function<br>
2.plot the fit<br>
<b>Data :<br>
Max </b>= 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25<br>
<b>Min</b> = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
month = np.arange(12)
plt.plot(month, temp_max, 'go') # go is green color
plt.plot(month, temp_min, 'co') #co is cyan color
plt.xlabel('Month')
plt.ylabel('Temperature')
```

```python
# 1. fitting it to a perodic function.
from scipy import optimize

def yearly_temp(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temp, month, temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temp, month, temp_min, [-40, 20, 0])
```

```python
# 2. plot graph.
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(month, temp_max, 'go')
plt.plot(days, yearly_temp(days, *res_max), 'm-')
plt.plot(month, temp_min, 'co')
plt.plot(days, yearly_temp(days, *res_min), 'y-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(url)
df
```

```python
# 1. Create a pie chart presenting the male/female proportion
count = df['sex'].value_counts()
colors = ['grey', 'orange']
```

```python
plt.pie(count.values.tolist(), labels = count.index.values.tolist(), colors = colors, startangle = 90, shadow=True, autopct = '%.1f%%')
plt.show()
```
